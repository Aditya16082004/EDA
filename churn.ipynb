## Data Cleaning

**1.** Create a copy of base data for manupulation & processing

```python
df2 =  df1.copy()
```

**2.** Total Charges should be numeric amount. Let's convert it to numerical data type

```python
df2.TotalCharges= pd.to_numeric(df2.TotalCharges,errors='coerce')
print(df2.isnull().sum())
```

**3.** As we can see there are 11 missing values in TotalCharges column. Let's check these records 

```python
df2.loc[df2 ['TotalCharges'].isnull() == True]
```

```python
#Removing missing values 
df2.dropna(how = 'any', inplace = True)

#df2.fillna(0)
```

**5.** Divide customers into bins based on tenure e.g. for tenure < 12 months: assign a tenure group if 1-12, for tenure between 1 to 2 Yrs, tenure group of 13-24; so on...

```python
# Get the max tenure
print(df2['tenure'].max()) #72
```

```python
# Group the tenure in bins of 12 months
labels = ["{0} - {1}".format(i, i + 11) for i in range(1, 72, 12)]

df2['tenure_group'] = pd.cut(df2.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
df2['tenure_group'].value_counts()
```

```python
#drop column customerID and tenure
df2.drop(columns= ['customerID','tenure'], axis=1, inplace=True)
df2.head()
```

```python
print(df2.isnull().sum())
```

### Univariate Analysis

```python

# Set a nice theme for all plots
sns.set_theme(style="whitegrid")

# Create a figure and adjust spacing
plt.figure(figsize=(15, 100))
plt.subplots_adjust(hspace=0.5)  # Increase the height space between plots

# Iterate through the columns and create count plots
for i, predictor in enumerate(df2.drop(columns=['Churn', 'TotalCharges', 'MonthlyCharges']).columns):
    plt.subplot(len(df2.columns) - 3, 1, i + 1)  # Adjust to the number of plots needed
    sns.countplot(data=df2, x=predictor, hue='Churn')
    plt.title(f'Count Plot of {predictor} by Churn')
    plt.xlabel(predictor)
    plt.ylabel('Count')
    plt.legend(title='Churn', loc='upper right')
    plt.xticks(rotation=45)
    plt.tight_layout()

plt.show()
```
