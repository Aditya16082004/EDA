```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the data
df = pd.read_csv('Loan 2.csv')
df
```

```python
df.isnull().sum()
```

```python
# Fill NaN in numeric columns with the median
numeric_cols = df.select_dtypes(include=[np.number]).columns
df[numeric_cols] = df[numeric_cols].apply(lambda x: x.fillna(x.median()))

# Fill NaN in categorical columns with the mode
categorical_cols = df.select_dtypes(include=[object]).columns
for col in categorical_cols:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)

# Identify and remove outliers using IQR method
Q1 = df[numeric_cols].quantile(0.25)
Q3 = df[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Filtering out the outliers
df_no_outliers = df[~((df[numeric_cols] < (Q1 - 1.5 * IQR)) | (df[numeric_cols] > (Q3 + 1.5 * IQR))).any(axis=1)]

print(df_no_outliers)
```

```python
df_no_outliers.isnull().sum()
```

```python
# EDA - Visualizations



# Scatter plot with outliers
plt.figure(figsize=(10, 5))
plt.scatter(df['Income (USD)'], df['Loan Amount Request (USD)'], alpha=0.5)
plt.title('Income vs Loan Amount Request (With Outliers)', color='white')
plt.xlabel('Income (USD)', color='white')
plt.ylabel('Loan Amount Request (USD)', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.savefig('scatter_plot_with_outliers.png', bbox_inches='tight', pad_inches=0.1)
plt.show()

# Scatter plot without outliers
plt.figure(figsize=(10, 5))
plt.scatter(df_no_outliers['Income (USD)'], df_no_outliers['Loan Amount Request (USD)'], alpha=0.5, color='orange')
plt.title('Income vs Loan Amount Request (Without Outliers)', color='white')
plt.xlabel('Income (USD)', color='white')
plt.ylabel('Loan Amount Request (USD)', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.savefig('scatter_plot_without_outliers.png', bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
# Distribution of Loan Amount Request
plt.figure(figsize=(10, 5))
sns.histplot(df_no_outliers['Loan Amount Request (USD)'], kde=True)
plt.title('Distribution of Loan Amount Request (USD)', color='white')
plt.xlabel('Loan Amount Request (USD)', color='white')
plt.ylabel('Frequency', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.savefig('loan_amount_distribution.png', bbox_inches='tight', pad_inches=0.1)
plt.show()

# Distribution of Loan Amount Request
plt.figure(figsize=(10, 5))
sns.histplot(df['Loan Amount Request (USD)'], kde=True)
plt.title('Distribution of Loan Amount Request (USD)', color='white')
plt.xlabel('Loan Amount Request (USD)', color='white')
plt.ylabel('Frequency', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.savefig('loan_amount_distribution.png', bbox_inches='tight', pad_inches=0.1)
plt.show()

```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
numeric_cols_no_outliers = df_no_outliers.select_dtypes(include=[np.number])
sns.heatmap(numeric_cols_no_outliers.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.savefig('correlation_heatmap.png', bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
# Boxplot for Loan Amount Request by Income Category (without outliers)
plt.figure(figsize=(10, 5))
sns.boxplot(x=df_no_outliers['Income (USD)'], y=df_no_outliers['Loan Amount Request (USD)'])
plt.title('Loan Amount Request by Income Category (Without Outliers)', color='white')
plt.xlabel('Income (USD)', color='white')
plt.ylabel('Loan Amount Request (USD)', color='white')
plt.xticks(color='white')
plt.yticks(color='white')
plt.savefig('boxplot.png', bbox_inches='tight', pad_inches=0.1)
plt.show()
```
